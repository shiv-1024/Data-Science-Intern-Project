```python
#Coverting Timestamp seconds to Datatime
bitcoin_df['timestamp'] = pd.to_datetime(bitcoin_df['timestamp'],unit = 's')

bitcoin_df.head()
```

```python
#Coverting Timestamp milliseconds to Datatime
trader_df['Timestamp'] = pd.to_datetime(trader_df['Timestamp'],unit='ms')

trader_df.head()
```

```python
#Removing Waste Features in Dataset
bitcoin_df = bitcoin_df.drop(columns=(['timestamp']),axis=0)
```

```python
bitcoin_df.head()
```

```python
#Taking the Date from the timestamp
trader_df['date'] = trader_df['Timestamp'].dt.floor('D')

trader_df.head()
```

```python
#coverting Date(str) to Datetime Datatype
bitcoin_df['date'] = pd.to_datetime(bitcoin_df['date'])
```

```python
#Merging Two Dataset into Single Dataset
merged_df = pd.merge(bitcoin_df,trader_df,on = 'date', how = 'inner')
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
#Coverting PNL value to Numeric Datatype
merged_df['Closed PnL'] = pd.to_numeric(merged_df['Closed PnL'], errors='coerce')
merged_df.isnull().sum()
```

```python
merged_df.tail()
```

```python
#PNl on Daily Basis
daily_pnl = merged_df.groupby(['Trade ID','date'])['Closed PnL'].sum().reset_index()
```

```python
daily_pnl.sample(5)
```

```python
daily_pnl.info()
```

```python
#win Rate Based on Trader 
win_rate = merged_df.groupby(["Trade ID"])["Closed PnL"].apply(lambda x : (x>0).mean()).reset_index()

win_rate.sample(10)
```

```python
#Avg Leverages Based on account
avg_trade_size = merged_df.groupby("Account")['Size USD'].mean().reset_index(name='avg_trade_size')

avg_trade_size.head()
```

```python
#metics for Total pnl,win Rate, Avg Trade Size and Total Trades

metrics = (
    merged_df.groupby("Account")
    .agg(
        total_pnl=("Closed PnL", "sum"),
        win_rate=("Closed PnL", lambda x: (x > 0).mean()),
        avg_trade_size=("Size USD", "mean"),
        total_trades=("Trade ID", "count")
    )
    .reset_index()
)

metrics.head()
```

```python
#Total trades Per day
trades_per_day = (merged_df.groupby('date')["Trade ID"].count().reset_index(name = 'num_trades'))
```

```python
trades_per_day.head()
```

```python
#Buying and Selling based on Account

buy_sell_per_trader = (merged_df.groupby("Account")["Side"].value_counts(normalize=True).unstack().reset_index())

buy_sell_per_trader
```

```python
#Total Perfomence with Total Pnl, avg Pnl , win rate and PnL voltality

total_perfomence = (
    merged_df.groupby('classification').agg(
    total_pnl = ("Closed PnL","sum"),
    avg_pnl = ("Closed PnL","mean"),
    win_rate = ("Closed PnL",lambda x : (x>0).mean()),
    pnl_voltality =("Closed PnL",'std')
)
.reset_index()
)

total_perfomence.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
plt.figure()
plt.bar(total_perfomence['classification'],total_perfomence['avg_pnl'])
plt.title("Average PnL: Fear vs Greed")
plt.xlabel("sentiment")
plt.ylabel("Average Pnl")
plt.show()
```

```python
win_rate = (
    merged_df.groupby("classification")['Closed PnL'].apply( lambda x : (x>0).mean()).reset_index(name = "win_rate")
    )


plt.figure()
sns.barplot(data=win_rate,x = "classification",y = "win_rate")
plt.title("win Rate: Fear vs Greed ")
plt.xlabel("sentiment")
plt.ylabel("Win Rate")
plt.show()
```

```python
plt.figure()
sns.barplot(data=total_perfomence,x='classification',y='pnl_voltality')
plt.title("Pnl Voltality Drawdown Proxy")
plt.xlabel("sentiment")
plt.ylabel("Voltality")
plt.show()
```

```python
trades_per_day = (
    merged_df.groupby(["classification", "date"])["Trade ID"]
    .count()
    .reset_index(name="daily_trades")
)

plt.figure()
sns.barplot(data=trades_per_day,x="classification",y="daily_trades")
plt.title("Trade Frequency per Day")
plt.xlabel("Sentiment")
plt.ylabel("Trades per Day")
plt.show()
```

```python
ave_position_size = merged_df.groupby("classification")['Size USD'].mean().reset_index(name = "avg_position_size")

plt.figure()
plt.bar(ave_position_size["classification"],ave_position_size["avg_position_size"])
plt.title("Average Position Size: Fear vs Greed")
plt.xlabel("Sentiment")
plt.ylabel("Avg Position Size (USD)")
plt.show()
```

```python
buy_sell = (
    merged_df.groupby("classification")["Side"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

plt.figure()
sns.barplot(data=buy_sell,x="classification",y="ratio",hue="Side")
plt.title("Long vs Short Ratio by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Ratio")
plt.show()
```

```python
trader_acitvity = merged_df.groupby("Account")["Trade ID"].count().reset_index(name="total_trades")

plt.figure()
sns.histplot(trader_acitvity['total_trades'],bins = 30)
plt.title("Frequent vs Infrequent Traders")
plt.xlabel("Total Trades")
plt.ylabel("Number of Traders")
plt.show()
```

```python
trader_win_rate = (merged_df.groupby("Account")["Closed PnL"].apply(lambda x: (x > 0).mean()).reset_index(name="win_rate"))

plt.figure()
sns.histplot(trader_win_rate["win_rate"], bins=20)
plt.title("Consistent vs Inconsistent Traders (Win Rate Distribution")
plt.xlabel("Win Rate")
plt.ylabel("Number of Traders")
plt.show()
```

```python
trader_volatility = (merged_df.groupby("Account")["Closed PnL"].std().reset_index(name="pnl_volatility"))

plt.figure()
sns.histplot(trader_volatility["pnl_volatility"], bins=20)
plt.title("PnL Volatility Distribution")
plt.xlabel("PnL Volatility")
plt.ylabel("Number of Traders")
plt.show()
```

```python
# If Fear means reduce Levarages size by 40%

merged_df["adjusted_size_usd"] = merged_df.apply(
    lambda row: row["Size USD"] * 0.6 
    if row["classification"] == "Fear"
    else row["Size USD"],
    axis=1
)
```
